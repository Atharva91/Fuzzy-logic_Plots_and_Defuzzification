```python
temp = np.arange(0,10,1)
def trapezoidal(temp,a,b,c,d):
    u = np.ndarray((len(temp),1))
    for i,x in enumerate((temp)):
        if(x<=a):
            u[i]=0
        elif(a<=x and x<=b):
            u[i]= ((x-a)/(b-a))
        elif(b<=x and x<=c):
            u[i]= 1
        elif(c<=x and x<=d):
            u[i]= (d-x)/(d-c)
        else:
            u[i]=0
    return u

temp = np.arange(0,10,1)
trapezoidal(temp,0,1,2,3)
first = trapezoidal(temp,*[0,1,4,5])
lim1 = len(first)
for i in range(lim1):
    if first[i] > 0.3:
        first[i] = 0.3
fig,a = plt.subplots()
a.plot(temp,first,'r')
a.set_ylabel('fuzzy Membership')
a.set_xlabel('temprature')
a.set_ylim(0,2)
```

```python
temp = np.arange(0,10,1)
def trapezoidal(temp,a,b,c,d):
    u = np.ndarray((len(temp),1))
    for i,x in enumerate((temp)):
        if(x<=a):
            u[i]=0
        elif(a<=x and x<=b):
            u[i]= ((x-a)/(b-a))
        elif(b<=x and x<=c):
            u[i]= 1
        elif(c<=x and x<=d):
            u[i]= (d-x)/(d-c)
        else:
            u[i]=0
    return u

temp = np.arange(0,10,1)
first = trapezoidal(temp,*[0,1,4,5])
lim1 = len(first)
for i in range(lim1):
    if first[i] > 0.3:
        first[i] = 0.3
        
second = trapezoidal(temp,*[3,4,6,7])
lim1 = len(second)
for i in range(lim1):
    if second[i] > 0.5:
        second[i] = 0.5

third = trapezoidal(temp,*[5,6,7,8])
lim1 = len(third)
for i in range(lim1):
    if third[i] > 1:
        third[i] = 1
        
fig,a=plt.subplots()
a.plot(temp,first,'r',temp,second,'b',temp,third,'r')
a.set_ylabel('Fuzzy Membership')
a.set_xlabel('X')





```

```python
max1 = np.fmax(first,second)
max2 = np.fmax(max1,third)
fig,a=plt.subplots()
a.plot(temp,max2,'g')
a.set_ylabel('Fuzzy Membership')
a.set_xlabel('X')
```

```python
# Centre of sum:- 

def Area(a,b,c,d,height):
    ar = height*0.5*((d-a)+(c-b))
    return (ar)

def Centroid(b,c):
    cent = (b+c)/2
    return(cent)

A1 = Area(0,1,4,5,0.3)
print(type(A1))
A2 = Area(3,4,6,7,0.5)
A3 = Area(5,6,7,8,1)

out_1 = Area(0,1,4,5,0.3)*Centroid(1,4)
print(type(out_1))
out_2 = Area(3,4,6,7,0.5)*Centroid(4,6)
out_3 = Area(5,6,7,8,1)*Centroid(6,7)

Final_out = (out_1+ out_2 + out_3) / (A1 + A2 + A3)
print(Final_out)


    
```
